```python
import json

with open('logs.json', 'r') as file:
    data = json.load(file)

data
```

```python
service_name = input("Enter name")
filtered_logs = [l for l in data["logs"] if l["service_sender"] == service_name]
filtered_logs
```

```python
service_sender = input("Enter sender: ")
service_reciever = input("Enter receiver: ")
filtered_logs = [l for l in data["logs"] if l["service_sender"] == service_sender and l["service_receiver"] == service_reciever]
filtered_logs
```
