```python
import sqlite3
from datetime import datetime

class Todo:
    def __init__(self, db_name="todo.db"):
        self.conn = sqlite3.connect(db_name)
        self.create_table()

    def create_table(self):
        query = """
        CREATE TABLE IF NOT EXISTS todos (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            title TEXT NOT NULL,
            description TEXT,
            create_date TEXT,
            expire_date TEXT
        );
        """
        self.conn.execute(query)
        self.conn.commit()

    def add_todo(self):
        n = int(input("How many todos do you want to insert? "))
        for i in range(n):
            title = input(f"Enter title for todo {i+1}: ")
            desc = input(f"Enter description for todo {i+1}: ")
            create_date = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            expire_date = input("Enter expire date (dd/mm/yyyy): ")
            self.conn.execute(
                "INSERT INTO todos (title, description, create_date, expire_date) VALUES (?, ?, ?, ?)",
                (title, desc, create_date, expire_date)
            )
        self.conn.commit()
        print("Todos inserted successfully.\n")

    def show_todos(self):
        cursor = self.conn.execute("SELECT id, title, description, create_date, expire_date FROM todos")
        rows = cursor.fetchall()
        if not rows:
            print("No todos found.\n")
        else:
            print("\n--- Todo List ---")
            for row in rows:
                print(f"{row[0]}. {row[1]} - {row[2]} - {row[3]} - {row[4]}")
            print()

    def update_todo(self):
        self.show_todos()
        try:
            todo_id = int(input("Enter the ID of the todo to update: "))
            new_title = input("Enter new title: ")
            new_desc = input("Enter new description: ")
            new_create_date = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            new_expire_date = input("Enter new expire date (dd/mm/yyyy): ")

            self.conn.execute(
                "UPDATE todos SET title = ?, description = ?, create_date = ?, expire_date = ? WHERE id = ?",
                (new_title, new_desc, new_create_date, new_expire_date, todo_id)
            )
            self.conn.commit()
            print("Todo updated successfully.\n")
        except ValueError:
            print("Invalid input.\n")

    def delete_todo(self):
        self.show_todos()
        try:
            todo_id = int(input("Enter the ID of the todo to delete: "))
            self.conn.execute("DELETE FROM todos WHERE id = ?", (todo_id,))
            self.conn.commit()
            print("Todo deleted successfully.\n")
        except ValueError:
            print("Invalid input.\n")

    def close(self):
        self.conn.close()


# Main Program
t = Todo()

while True:
    print(""" 
1. Insert todo
2. Update todo
3. Display todo
4. Delete todo
5. Quit program
""")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if choice == 1:
        t.add_todo()
    elif choice == 2:
        t.update_todo()
    elif choice == 3:
        t.show_todos()
    elif choice == 4:
        t.delete_todo()
    elif choice == 5:
        print("Quit")
        t.close()
        break
    else:
        print("Invalid choice. Please try again.")
```
